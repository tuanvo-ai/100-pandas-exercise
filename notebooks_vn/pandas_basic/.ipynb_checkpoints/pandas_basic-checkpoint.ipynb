### Dữ liệu bị thiếu

```python
# Đọc dữ liệu product
product_path = "../../data/preprocessed/product.csv"

dtype = {
    'product_cd': str,
    'category_major_cd': int,
    'category_medium_cd': int,
    'category_small_cd': int,
    'unit_price': float,
    'unit_cost': float
}
df_product = pd.read_csv(product_path, dtype=dtype)
print(df_product.shape)
df_product.head()
```

```python
df_product.isna().sum()
```

```python
df_product[df_product.isna().any(axis=1)]
```

```python
# Sử dụng .fill() để điền vào giá trị bị thiếu
unit_price_mean = df_product["unit_price"].mean()
print(f"Mean unit price: {unit_price_mean}")

df_product.loc[:, "unit_price"] = df_product["unit_price"].fillna(unit_price_mean)
```

### Kết hợp nhiều dữ liệu

#### concat()

```python
df_tmp_1 = df_store.query("prefecture_cd == 12")
df_tmp_2 = df_store.query("prefecture_cd == 14")
```

```python
df_tmp_2
```

```python
# Nối các bảng theo từng hàng với nhau
pd.concat([df_tmp_1, df_tmp_2])
```

#### merge()

```python
# sử dụng merge() để join 2 hoặc nhiều bảng dữ liệu
left = pd.DataFrame({"key": ["A", "B", "C"], "lval": [1, 2, 99]})
display(left)

right = pd.DataFrame({"key": ["A", "B", "F"], "rval": [4, 5, 6]})
display(right)
```

```python
# Inner Join
pd.merge(left, right, on="key", how="inner")
```

```python
# Left Join
pd.merge(left, right, on="key", how="left")
```

```python
# Right Join
pd.merge(left, right, on="key", how="right")
```

```python
# Outter Join
pd.merge(left, right, on="key", how="outer")
```

### Pivot table

```python
import numpy as np

df = pd.DataFrame(
    {
        "A": ["A", "B", "C"] * 4,
        "B": ["Tokyo", "Chiaba"] * 6,
        "C": np.random.randn(12),
        "D": np.arange(12),
    }
)

pd.pivot_table(df, values=["C", "D"], index=["A"], columns=["B"])
```

### Chuỗi thời gian

```python
# Tạo một chuỗi thời gian
date_range = [f"2024080{str(i)}" for i in np.arange(1,7)]

# Tạo DataFrame với chuỗi thời gian
df = pd.DataFrame({'date': date_range, 'value': range(6)})
print(df)
```

```python
# Chuyển đổi cột 'date' thành định dạng datetime
df['date'] = pd.to_datetime(df['date'], format="%Y%m%d")
print(df.dtypes)
df
```

```python
# Trích xuất năm, tháng, ngày từ cột 'date'
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
display(df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự theo định dạng 'YYYY-MM-DD'
df['formatted_date'] = df['date'].dt.strftime('%Y/%m/%d')
display(df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự với các thành phần khác nhau
df['formatted_date_full'] = df['date'].dt.strftime('%A, %d %B %Y')
display(df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự theo định dạng tùy chỉnh 'DD/MM/YYYY'
df['custom_formatted_date'] = df['date'].dt.strftime('%d/%m/%Y')
display(df)
```

**Các định dạng thông dụng sử dụng trong strftime:**
* `%Y`: Năm đầy đủ (ví dụ: 2024)
* `%m`: Tháng (01 đến 12)
* `%d`: Ngày (01 đến 31)
* `%H`: Giờ (00 đến 23)
* `%M`: Phút (00 đến 59)
* `%S`: Giây (00 đến 59)
* `%A`: Tên đầy đủ của ngày trong tuần (ví dụ: Monday)
* `%B`: Tên đầy đủ của tháng (ví dụ: January)
