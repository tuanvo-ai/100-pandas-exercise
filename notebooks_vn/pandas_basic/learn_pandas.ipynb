# chuỗi thời gian

```python
# Tạo một chuỗi thời gian
data_range = [f"2024100{str(i)}" for i in np.arange(1,7)]

# Tạo DataFrame với chuỗi thời gian
df = pd.DataFrame({'date': data_range, 'value': range(6)})
print(df)
```

```python
# Chuyển đổi cột 'date' thành định dạng datetime
#df['date'] = pd.to_datetime(df['date'], format="%Y%m%d")
df['date'] = pd.to_datetime(df['date'], format="%Y%m%d")
print(df.dtypes)
df
```

```python
# Trích xuất năm, tháng, ngày từ cột 'date'
df['year']= df['date'].dt.year
df['month']= df['date'].dt.month
df['day']= df['date'].dt.day

display(df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự theo định dạng 'YYYY-MM-DD'
df['format_date'] = df['date'].dt.strftime('%d/%m/%Y')

#df['formatted_date'] = df['date'].dt.strftime('%Y/%m/%d')
display(df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự với các thành phần khác nhau
df['format_date_full'] = df['date'].dt.strftime('%A, %d %B %Y')
#df['formatted_date_full'] = df['date'].dt.strftime('%A, %d %B %Y')
display(df)
```
