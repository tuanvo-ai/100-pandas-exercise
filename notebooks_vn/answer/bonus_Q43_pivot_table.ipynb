# Giải thích pivot table

```python
import pandas as pd

# Tạo DataFrame
data = {
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar'],
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Sales': [150, 200, 250, 300, 100, 400, 500, 300, 600]
}

df = pd.DataFrame(data)

print("DataFrame ban đầu:")
display(df)

# Tạo pivot table
pivot_df = pd.pivot_table(df, values='Sales', index='Month', columns='Product', aggfunc='sum')

print("\nPivot Table:")
display(pivot_df)
```
